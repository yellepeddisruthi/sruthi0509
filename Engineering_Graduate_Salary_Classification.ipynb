```python

category_mapping = {
    'computer science & engineering': 'Computer Science & IT',
    'information technology': 'Computer Science & IT',
    'electronics & telecommunications': 'Electronics & Communication',
    'electronics and communication engineering': 'Electronics & Communication',
    'mechanical engineering': 'Mechanical & Production',
    'civil engineering': 'Other',
    'chemical engineering': 'Other',
    'biotechnology': 'Other',
    'electrical engineering': 'Other',

}

# Map the old categories to the new categories
data['Specialization_Category'] = data['Specialization'].map(category_mapping)

# Handle any missing categories that are not in the mapping
data['Specialization_Category'].fillna('Other', inplace=True)

# Display the updated dataset with new category
print(data[['Specialization', 'Specialization_Category']].head())
```

```python
print(data['Specialization_Category'].value_counts())
```

```python
# Step 1: Calculate the median of the Salary column
median_salary = data['Salary'].median()

# Step 2: Create a new column with values 1 and 2 based on the median salary
data['SalaryCategory_Median'] = data['Salary'].apply(lambda x: 0 if x >= median_salary else 1)

# Display the first few rows of the updated dataset with the new median-based category
data[['Salary', 'SalaryCategory_Median']].head()
```

```python
df= data.replace(to_replace='m', value='1')
df= df.replace(to_replace='f', value='0')
df
```

```python
df.describe()
```

```python
X=df.drop(['SalaryCategory_Median'],axis=1)

print(X)
y = df['SalaryCategory_Median']
print(y)
```

```python
X= pd.get_dummies(X, drop_first=True)
print(X)
```

```python
X = X.applymap(lambda x: 1 if x is True else (0 if x is False else x))
print(X)
```

# **VIF**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i).round(1) for i in range(X.shape[1])]

    return(vif)

calc_vif(X)
```
